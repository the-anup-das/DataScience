```
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```
df=pd.read_csv('Titanic_Detailed.csv')
df.head()
```

```
#1. Make a distribution plot (histogram) of the 'Fare' attribute of the data using Matplotlib

plt.hist(df['Fare'])
plt.ylabel('Fare')
plt.show()

plt.hist('Fare', data=df) 
```

```
#2. Make a distribution plot of the 'Fare' attribute of the data using Seaborn
sns.histplot(df['Fare'], bins=16, color="purple")

plt.hist('Fare', data=df) 
```

```
#3. Make a joint plot comparing the 'Fare' and 'Age' attributes of the data using Seaborn.
sns.jointplot(x='Fare',y='Age',data=df,kind='hex')
```

```
#4. Make a box plot comparing the 'Pclass' attribute (x axis) with the 'Age' attribute (y axis) using Seaborn
sns.boxplot( x=df["Pclass"], y=df["Age"]);
plt.show()

sns.boxplot(x='Pclass', y='Age', data=df) 
```

```
#5. Make a violin plot comparing the 'Embarked' attribute (x axis) with the 'Age' attribute (y axis) using Seaborn
fig=px.violin(df,x='Embarked',y='Age')
fig.show()

sns.violinplot(x='Embarked',y='Age',data=df) 
```

```
#6. Make a count plot of the 'Pclass' attribute using Seaborn to display the number of passengers in each class
sns.countplot(data=df,x='Pclass')
```

```
#7. Calculate the correlation of the DataFrame using Pandas, and make a heatmap of the correlation using Plotly (Don't forget to add the labels of the dataframe to the x and y axis).
corr=df.corr()
sns.heatmap(corr)

px.imshow(corr) 
```
